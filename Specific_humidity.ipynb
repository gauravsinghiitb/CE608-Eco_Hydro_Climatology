```python
import xarray as xr
import glob
import cfgrib

# Ensure cfgrib is installed
try:
    import cfgrib
except ImportError:
    raise ImportError("cfgrib is not installed. Install it using: pip install cfgrib")



# Step 1: Define directory path
grib_dir = "/content/data"

# Step 2: Get all GRIB files in the directory
grib_files = sorted(glob.glob(grib_dir + "/sh_*.grib"))

# Step 3: Debugging print statements
print("Found GRIB files:", grib_files)

# Step 4: Open files and check time dimension
ds_list = []
for file in grib_files:
    try:
        ds = xr.open_dataset(file, engine="cfgrib")

        # Print available variables and dimensions
        print(f"Reading {file}:")
        print("Variables:", list(ds.variables))
        print("Dimensions:", ds.dims)

        if "time" not in ds.dims:
            print(f"Warning: 'time' dimension missing in {file}. Skipping.")
        else:
            ds_list.append(ds)
    except Exception as e:
        print(f"Error reading {file}: {e}")

# Step 5: Merge datasets if any were successfully opened
if ds_list:
    ds = xr.concat(ds_list, dim="time", combine_attrs="override")
    print("Merged dataset is ready.")
else:
    print("No valid datasets found. Check file paths and content.")
```

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Ensure 'valid_time' is a proper datetime object
ds["time"] = xr.decode_cf(ds)["time"]

# Extract the year from the time dimension
ds = ds.assign_coords(year=ds["time"].dt.year)

# Compute annual max and min specific humidity (spatially averaged)
annual_max_humidity = ds.q.groupby("year").max(dim="time").mean(dim=["latitude", "longitude"])
annual_min_humidity = ds.q.groupby("year").min(dim="time").mean(dim=["latitude", "longitude"])

# Convert year coordinate to NumPy array
years = annual_max_humidity.coords["year"].values

# **Fix: Remove extra dimensions (squeeze)**
annual_max_humidity = annual_max_humidity.squeeze()
annual_min_humidity = annual_min_humidity.squeeze()

# Ensure they are 1D before regression
if annual_max_humidity.ndim > 1:
    annual_max_humidity = annual_max_humidity.mean(dim="step")  # Adjust if needed
if annual_min_humidity.ndim > 1:
    annual_min_humidity = annual_min_humidity.mean(dim="step")  # Adjust if needed

# Perform linear regression for trends
slope_max, intercept_max, r_value_max, p_value_max, std_err_max = linregress(years, annual_max_humidity.values)
slope_min, intercept_min, r_value_min, p_value_min, std_err_min = linregress(years, annual_min_humidity.values)

# Plot the trend for max specific humidity
plt.figure(figsize=(8, 5))
plt.plot(years, annual_max_humidity, marker='o', linestyle='-', label="Annual Max Specific Humidity")
plt.plot(years, slope_max * years + intercept_max, linestyle="dashed", color="red", label=f"Trend (p={p_value_max:.3f})")
plt.xlabel("Year")
plt.ylabel("Specific Humidity")
plt.title("Annual Maximum Specific Humidity Trend (1990-2025)")
plt.legend()
plt.grid()
plt.show()

# Plot the trend for min specific humidity
plt.figure(figsize=(8, 5))
plt.plot(years, annual_min_humidity, marker='o', linestyle='-', label="Annual Min Specific Humidity", color="blue")
plt.plot(years, slope_min * years + intercept_min, linestyle="dashed", color="red", label=f"Trend (p={p_value_min:.3f})")
plt.xlabel("Year")
plt.ylabel("Specific Humidity")
plt.title("Annual Minimum Specific Humidity Trend (1990-2025)")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
from scipy.interpolate import griddata

# Extract data
q = ds["q"].values  # Shape: (time, latitude, longitude)
time = ds["time"].values
lat = ds["latitude"].values
lon = ds["longitude"].values

# Convert time to years
time = np.array([t.astype('datetime64[Y]').astype(int) + 1970 for t in time])

# Prepare array for storing slopes (specific humidity trend at each grid point)
trend_map = np.full((len(lat), len(lon)), np.nan)

# Compute trends for each grid cell
for i in range(len(lat)):
    for j in range(len(lon)):
        humidity_series = q[:, i, j]  # Extract time series for the grid cell

        if np.all(np.isnan(humidity_series)):  # Skip if all values are NaN
            continue

        slope, _, _, p_value, _ = linregress(time, humidity_series)

        # Only consider significant trends (p-value < 0.05)
        if p_value < 0.05:
            trend_map[i, j] = slope

# Interpolate missing values
lat_grid, lon_grid = np.meshgrid(lat, lon, indexing='ij')
mask = ~np.isnan(trend_map)
points = np.column_stack((lat_grid[mask], lon_grid[mask]))
values = trend_map[mask]
trend_map_interp = griddata(points, values, (lat_grid, lon_grid), method='linear')

# Plot the trend map
plt.figure(figsize=(10, 5))
plt.pcolormesh(lon, lat, trend_map_interp, cmap="coolwarm", shading="nearest")
plt.colorbar(label="Specific Humidity Trend (kg/kg per year)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Spatial Trend in Specific Humidity (1990-2025)")
plt.show()
```
